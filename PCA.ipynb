```python
from sklearn.datasets import load_iris
iris=load_iris()
iris_data=iris.data
iris_data.shape
```

```python
iris_labels=iris.target
iris_labels.shape
```

```python
import numpy as np
labels=np.reshape(iris_labels,(150,1))
```

```python
final_iris_data=np.concatenate([iris_data,labels],axis=1)
final_iris_data.shape
```

```python
import pandas as pd
iris_dataset=pd.DataFrame(final_iris_data)
features=iris.feature_names
features
```

```python
features_label=np.append(features,'label')
```

```python
iris_dataset.columns=features_label
iris_dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(iris_dataset)
```

```python
scaled_data=scaler.transform(iris_dataset)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
scaled_data
```

```python
x_pca
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=iris['target'])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```
